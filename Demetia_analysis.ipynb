```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create dementia data manually
data = {
    'Country': ['China', 'India', 'USA', 'Brazil', 'Indonesia', 'Russia', 'Mexico', 'Japan', 'Germany', 'UK'],
    'Number of People with Dementia (millions)': [14, 8.8, 6.5, 4.1, 3.5, 2.7, 2.4, 2.2, 1.9, 1.6]
}

df = pd.DataFrame(data)

# Show the data
print(df)

# Plot the data
plt.figure(figsize=(10,6))
sns.barplot(x='Number of People with Dementia (millions)', y='Country', data=df, palette='magma')
plt.title('Top 10 Countries with Highest Dementia Cases')
plt.xlabel('People with Dementia (millions)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
